```
%sql
select CANAL  from youtube
```

```
%sql
select CANAL,
CONCAT(SUBSTR(INSCRITOS,1,1),'.',SUBSTR(INSCRITOS,2,3),'.',SUBSTR(INSCRITOS,4,3)) as INSCRITOS,
CONCAT(SUBSTR(VIDEOS,1,1),'.',SUBSTR(VIDEOS,2,3)) as VIDEOS,
CONCAT(SUBSTR(VIEWS_CANAL,0,3),'.',SUBSTR(VIEWS_CANAL,3,3),'.',SUBSTR(VIEWS_CANAL,6,3)) as VIEWS_CANAL
 from (
select CANAL,INSCRITOS ,VIDEOS,VIEWS_CANAL  from youtube
group by CANAL,INSCRITOS,VIDEOS,VIEWS_CANAL)
```

```
%sql
--select * from youtube_table

select 
CONCAT(SUBSTR(VIS_PODCAST,1,1),'.',SUBSTR(VIS_PODCAST,2,3),'.',SUBSTR(VIS_PODCAST,4,3)) as VIS_PODCAST,
CONCAT(SUBSTR(LIKES_PODCAST,1,3),'.',SUBSTR(LIKES_PODCAST,4,4)) as LIKES_PODCAST,
CONCAT(SUBSTR(COMENT_PODCAST,1,1),'.',SUBSTR(COMENT_PODCAST,2,3)) as COMENT_PODCAST
from (
    select sum(VISUALIZACAO) as VIS_PODCAST, sum(LIKES) as LIKES_PODCAST,  sum(COMENARIOS) as COMENT_PODCAST from youtube)

```
