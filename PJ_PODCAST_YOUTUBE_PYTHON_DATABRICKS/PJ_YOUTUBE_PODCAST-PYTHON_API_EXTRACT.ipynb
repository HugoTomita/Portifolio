```python
#Coletando info da playlist
playlist_videos=[] #Criando lista vazia
res = youtube.playlistItems().list(part='snippet', playlistId = playlistId, maxResults=50).execute() # Chamando api para capturar dados
playlist_videos= res['items']

# Se houver mais de 50 videos, usar o loop
#while True:
#    res = youtube.playlistItems().list(part='snippet', playlistId = playlistId,maxResults=50,pageToken=nextPage_token ).execute() 
#    playlist_videos += res['items']
#    
#    nextPage_token  = res.get('nestPageToken')
#    
#    if nextPage_token is None:
#        break

```

```python
videos_ids=list(map(lambda x: x['snippet']['resourceId']['videoId'],playlist_videos))
videos_ids
```

```python
#coletando info dos videos
stats=[] 

for videos_ids in videos_ids:
    res = youtube.videos().list(part='statistics', id=videos_ids).execute()
    stats+=res['items']

stats
```

```python
YT_canal=list(map(lambda x: x['snippet']['channelTitle'],playlist_videos))
YT_titulo=list(map(lambda x: x['snippet']['title'],playlist_videos))
YT_url=list(map(lambda x: x['snippet']['thumbnails'],playlist_videos))
YT_data=list(map(lambda x: x['snippet']['publishedAt'],playlist_videos))
YT_descricao=list(map(lambda x: x['snippet']['description'],playlist_videos))
YT_canalid=list(map(lambda x: x['snippet']['channelId'],playlist_videos))
YT_videoid=list(map(lambda x: x['snippet']['resourceId']['videoId'],playlist_videos))
YT_visualizacao=list(map(lambda x: x['statistics']['viewCount'],stats))
YT_likes=list(map(lambda x: x['statistics']['likeCount'],stats))
YT_comentarios=list(map(lambda x: x['statistics']['commentCount'],stats))
```

```python
extraction_date = [str(datetime.now())]*len(videos_ids)
```

```python
#Criadno o DataFrame com as informacoes da playlist do podcast
df_playlist = pd.DataFrame({
    'CANAL':YT_canal,
    'TITULO':YT_titulo,
    'URL':YT_url,
    'DATA':YT_data,
    'DESCRICAO':YT_descricao,
    'IDCANAL':YT_canalid,
    'ID VIDEO':YT_videoid,
    'VISUALIZACAO':YT_visualizacao,
    'LIKES': YT_likes,
    'COMENARIOS':YT_comentarios})
df_playlist.head()
```
